### Addition using ML regression model

```python
import pandas as pd
```

```python
data = pd.read_csv('add.csv')
data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(data['x'],data['sum'])
```

```python
plt.scatter(data['y'],data['sum'])
```

```python
#  Store Feature Matrix In X And Response (Target) In Vector y
X  = data[['x','y']]
y= data['sum']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.33,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#Check Model’s Prediction Performance
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
# Comparing The Results
y_pred = model.predict(X_test)
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

```python
# Actual Prediction
model.predict([[10,20]])
```

```python
# Save Model Usign Joblib
import joblib
joblib.dump(model,'model_joblib_additionof2numbers')
```

```python
model = joblib.load('model_joblib_additionof2numbers')
model.predict([[30,40]])
```

#### GUI

```python
from tkinter import *
import joblib
```

```python
def show_entry_fields():
    num1 = int(e1.get())
    num2 = int(e2.get())
    model = joblib.load('model_joblib_additionof2numbers')
    result=model.predict([[num1,num2]])
    print("Sum is : ",result)
    label_result = Label(master, text="Sum is : "+str(result))
    label_result.grid(row=4, column=1)
```

```python
master = Tk()
master.title("Addition of Two Numbers Using ML")
label = Label(master, text="Enter Two Numbers", bg= "light blue", fg= "black")
label.grid(row=0, column=1)
e1 = Entry(master)
e1.grid(row=1, column=0)
e2 = Entry(master)
e2.grid(row=1, column=2)
button = Button(master, text="Predict", command=show_entry_fields)
button.grid(row=2, column=1)
master.mainloop()
```
